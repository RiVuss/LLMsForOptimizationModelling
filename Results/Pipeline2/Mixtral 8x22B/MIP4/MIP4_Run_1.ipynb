```python
import pyomo.environ as pyo
import numpy as np

# Step 1: Define parameters and variables

T = 10  # Total number of time steps
P = 12  # Total number of available pipes
A = range(1, 11)  # Set of activated pipes

# Sample data for capacities of pipes
C = {p: np.random.uniform(1, 5) for p in A}

# Inflow function at point A
f = lambda t: max(-4/30 * t**3 + t**2 - 0.234 * t + 3, 0)

# Create Pyomo model
model = pyo.AbstractModel()

# Define sets
model.P = pyo.Set(initialize=range(1, P + 1))
model.A = pyo.Set(initialize=A)
model.T = pyo.Set(initialize=range(1, T + 1))
model.N = pyo.Set(initialize=['A', 'B', 'C', 'D', 'E', 'F', 'G'])

# Define parameters
model.C = pyo.Param(model.A, initialize=C)

# Define variables
model.x = pyo.Var(model.A, model.T, bounds=(0, None))
model.y = pyo.Var(model.P, within=pyo.Binary)
model.z = pyo.Var(model.N, model.T, bounds=(0, None))

# Step 2: Define the objective function

def objective_rule(model):
    return sum(model.x[('EG', t)] + model.x[('FG', t)] for t in model.T)

model.objective = pyo.Objective(rule=objective_rule, sense=pyo.maximize)

# Step 3: Define the constraints

def flow_conservation_rule(model, n, t):
    in_pipes = []
    out_pipes = []

    if n == 'A':
        out_pipes = [p for p in model.A if p[0] == 'A']
        return sum(model.x[p, t] for p in out_pipes) == f(t)
    elif n == 'G':
        in_pipes = [p for p in model.A if p[1] == 'G']
        return sum(model.x[p, t] for p in in_pipes) == model.z[n, t]
    else:
        in_pipes = [p for p in model.A if p[1] == n]
        out_pipes = [p for p in model.A if p[0] == n]
        return sum(model.x[p, t] for p in in_pipes) + model.z[n, t] == sum(model.x[p, t] for p in out_pipes)

model.flow_conservation = pyo.Constraint(model.N, model.T, rule=flow_conservation_rule)

def capacity_rule(model, p, t):
    return model.x[p, t] <= model.C[p] * model.y[p]

model.capacity = pyo.Constraint(model.A, model.T, rule=capacity_rule)

def activation_rule(model):
    return sum(model.y[p] for p in model.P) == 10

model.activation = pyo.Constraint(rule=activation_rule)

# Step 4: Solve the optimization problem

# Create a concrete instance of the model
instance = model.create_instance()

# Solve the instance
solver = pyo.SolverFactory('glpk')
results = solver.solve(instance)

# Print the results
print(results)
print("Optimal Flow to Point G:", pyo.value(instance.objective.expr))
```
