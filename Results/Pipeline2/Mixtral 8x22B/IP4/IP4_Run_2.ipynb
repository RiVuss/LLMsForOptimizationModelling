```python
import pyomo.environ as pyo

# Define the model
model = pyo.AbstractModel()

# Define the parameters
model.S = pyo.Param(initialize=5500)  # Sales target
model.H_f = pyo.Param(initialize=800)  # Full-time employee hours target
model.H_p = pyo.Param(initialize=320)  # Part-time employee hours target
model.O_f = pyo.Param(initialize=100)  # Full-time employee overtime target
model.R_f = pyo.Param(initialize=5)  # Full-time employee sales rate
model.R_p = pyo.Param(initialize=2)  # Part-time employee sales rate
model.P1 = pyo.Param(initialize=1)  # Priority factor for sales deviation
model.P2 = pyo.Param(initialize=1)  # Priority factor for full-time employee overtime deviation
model.P3 = pyo.Param(initialize=2)  # Priority factor for full-time employee hours deviation
model.P4 = pyo.Param(initialize=3)  # Priority factor for part-time employee hours and overtime deviation

# Define the variables
model.x_s = pyo.Var()  # Actual sales
model.x_h_f = pyo.Var()  # Actual full-time employee hours
model.x_h_p = pyo.Var()  # Actual part-time employee hours
model.x_o_f = pyo.Var()  # Actual full-time employee overtime hours
model.x_o_p = pyo.Var()  # Actual part-time employee overtime hours
model.d_s = pyo.Var()  # Sales deviation
model.d_h_f = pyo.Var()  # Full-time employee hours deviation
model.d_h_p = pyo.Var()  # Part-time employee hours deviation
model.d_o_f = pyo.Var()  # Full-time employee overtime deviation
model.d_o_p = pyo.Var()  # Part-time employee overtime deviation

# Objective function
def obj_rule(model):
    return model.P1 * abs(model.d_s) + model.P2 * abs(model.d_o_f) + model.P3 * abs(model.d_h_f) + model.P4 * (model.x_h_p + 3 * model.x_o_p)
model.obj = pyo.Objective(rule=obj_rule, sense=pyo.minimize)

# Constraints
def sales_constraint_rule(model):
    return model.x_s == model.S + model.d_s
model.sales_constraint = pyo.Constraint(rule=sales_constraint_rule)

def full_time_hours_constraint_rule(model):
    return model.x_h_f == model.H_f + model.d_h_f
model.full_time_hours_constraint = pyo.Constraint(rule=full_time_hours_constraint_rule)

def part_time_hours_constraint_rule(model):
    return model.x_h_p == model.H_p + model.d_h_p
model.part_time_hours_constraint = pyo.Constraint(rule=part_time_hours_constraint_rule)

def full_time_overtime_constraint_rule(model):
    return model.x_o_f == model.O_f + model.d_o_f
model.full_time_overtime_constraint = pyo.Constraint(rule=full_time_overtime_constraint_rule)

def sales_equation_rule(model):
    return model.x_s == (model.x_h_f * model.R_f) + (model.x_h_p * model.R_p)
model.sales_equation = pyo.Constraint(rule=sales_equation_rule)

def full_time_hours_limit_rule(model):
    return model.x_h_f <= model.H_f + model.O_f
model.full_time_hours_limit = pyo.Constraint(rule=full_time_hours_limit_rule)

def part_time_hours_limit_rule(model):
    return model.x_h_p <= model.H_p + model.x_o_p
model.part_time_hours_limit = pyo.Constraint(rule=part_time_hours_limit_rule)

def full_time_overtime_limit_rule(model):
    return model.x_o_f <= model.O_f
model.full_time_overtime_limit = pyo.Constraint(rule=full_time_overtime_limit_rule)

def non_negativity_rule(model):
    return [model.x_s >= 0, model.x_h_f >= 0, model.x_h_p >= 0, model.x_o_f >= 0, model.x_o_p >= 0, model.d_s >= 0, model.d_h_f >= 0, model.d_h_p >= 0, model.d_o_f >= 0, model.d_o_p >= 0]
model.non_negativity = pyo.Constraint(rule=non_negativity_rule)

# Solve the model
solver = pyo.SolverFactory('glpk')
solver.solve(model)
```
