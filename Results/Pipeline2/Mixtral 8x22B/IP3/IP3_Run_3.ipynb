### Code Executability

```python
import pyomo.environ as pyo

# Sample data
I = ['item1', 'item2', 'item3', 'item4']  # Set of items
B = ['box1', 'box2', 'box3']  # Set of boxes
W = {'item1': 5, 'item2': 3, 'item3': 7, 'item4': 2}  # Weight of items
C = {'box1': 10, 'box2': 8, 'box3': 12}  # Capacity of boxes
M = 2  # Upper bound on the number of boxes

model = pyo.ConcreteModel()

# Sets
model.I = pyo.Set(initialize=I)
model.B = pyo.Set(initialize=B)

# Parameters
model.W = pyo.Param(model.I, initialize=W)
model.C = pyo.Param(model.B, initialize=C)
model.M = pyo.Param(initialize=M)

# Decision variables
model.x = pyo.Var(model.I, model.B, domain=pyo.Binary)
model.y = pyo.Var(model.B, domain=pyo.Binary)

# Objective function
def obj_func(model):
    return pyo.summation(model.y)

model.obj = pyo.Objective(rule=obj_func, sense=pyo.minimize)

# Constraints
def item_constraint(model, i):
    return pyo.summation(model.x[i, b] for b in model.B) == 1

model.item_c = pyo.Constraint(model.I, rule=item_constraint)

def capacity_constraint(model, b):
    return pyo.summation(model.W[i] * model.x[i, b] for i in model.I) <= model.C[b] * model.y[b]

model.capacity_c = pyo.Constraint(model.B, rule=capacity_constraint)

def max_boxes_constraint(model):
    return pyo.summation(model.y) <= model.M

model.max_boxes_c = pyo.Constraint(rule=max_boxes_constraint)

# Solve the optimization problem
solver = pyo.SolverFactory('glpk')
solver.solve(model)

# Print the solution
print("Objective value:", pyo.value(model.obj))
for b in model.B:
    print(f"Box {b}: y = {pyo.value(model.y[b])}")
    for i in model.I:
        print(f"  Item {i}: x = {pyo.value(model.x[i, b])}")
```

### Solution Correctness

```python
import pyomo.environ as pyo


B = 9
W = [2,3,4,5,6,7,8]
q = [4,2,6,6,2,2,2]
s=[]
for j in range(len(W)):
    for i in range(q[j]):
        s.append(W[j])


def FFD(s, B):
    remain = [B]
    sol = [[]]
    for item in sorted(s, reverse=True):
        for j,free in enumerate(remain):
            if free >= item:
                remain[j] -= item
                sol[j].append(item)
                break
        else:
            sol.append([item])
            remain.append(B-item)
    return sol
N = len(s)
M = len(FFD(s, B))

model = pyo.ConcreteModel()

# Sets
model.I = pyo.Set(initialize=range(N))
model.B = pyo.Set(initialize=range(M))

# Parameters
model.W = pyo.Param(model.I, initialize=s)
model.C = pyo.Param(model.B, initialize=[B] * M)
model.M = pyo.Param(initialize=M)

# Decision variables
model.x = pyo.Var(model.I, model.B, domain=pyo.Binary)
model.y = pyo.Var(model.B, domain=pyo.Binary)

# Objective function
def obj_func(model):
    return sum(model.y[b] for b in model.B)

model.obj = pyo.Objective(rule=obj_func, sense=pyo.minimize)

# Constraints
def item_constraint(model, i):
    return sum(model.x[i, b] for b in model.B) == 1

model.item_c = pyo.Constraint(model.I, rule=item_constraint)

def capacity_constraint(model, b):
    return sum(model.W[i] * model.x[i, b] for i in model.I) <= model.C[b] * model.y[b]

model.capacity_c = pyo.Constraint(model.B, rule=capacity_constraint)

def max_boxes_constraint(model):
    return sum(model.y[b] for b in model.B) <= model.M

model.max_boxes_c = pyo.Constraint(rule=max_boxes_constraint)

# Solve the optimization problem
solver = pyo.SolverFactory('glpk')
solver.solve(model)

# Print the solution
print("Objective value:", pyo.value(model.obj))
for b in model.B:
    print(f"Box {b}: y = {pyo.value(model.y[b])}")
    for i in model.I:
        print(f"  Item {i}: x = {pyo.value(model.x[i, b])}")
```
