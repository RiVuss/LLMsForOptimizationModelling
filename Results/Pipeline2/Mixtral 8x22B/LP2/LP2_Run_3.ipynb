### Code Executability

```python
import pyomo.environ as pyo

# STEP 1: Define parameters and variables

# Define the model
model = pyo.ConcreteModel()

# Define the parameters
B = 100000  # budget
C1 = 10  # cost per unit of raw material 1
C2 = 20  # cost per unit of raw material 2
Q1 = 5000  # max storage capacity for raw material 1
Q2 = 3000  # max storage capacity for raw material 2
R1 = 2000  # required amount of active ingredient for drug 1
R2 = 3000  # required amount of active ingredient for drug 2
P1 = 50  # operational cost for producing one unit of drug 1
P2 = 70  # operational cost for producing one unit of drug 2
S1 = 200  # selling price per unit of drug 1
S2 = 300  # selling price per unit of drug 2
M1 = 500  # manpower and equipment hours available for drug 1
M2 = 700  # manpower and equipment hours available for drug 2
T1 = 10  # time required to produce one unit of drug 1
T2 = 15  # time required to produce one unit of drug 2
E1 = 0.995  # extraction efficiency for active ingredient from raw material 1
E2 = 0.98  # extraction efficiency for active ingredient from raw material 2

# Define the variables
model.x1 = pyo.Var(within=pyo.NonNegativeReals)  # quantity of raw material 1
model.x2 = pyo.Var(within=pyo.NonNegativeReals)  # quantity of raw material 2
model.y1 = pyo.Var(within=pyo.NonNegativeReals)  # amount of drug 1 produced
model.y2 = pyo.Var(within=pyo.NonNegativeReals)  # amount of drug 2 produced

# STEP 2: Define the objective function

# Define the objective function
def obj_function(model):
    return (S1 * model.y1 + S2 * model.y2) - (C1 * model.x1 + C2 * model.x2 + P1 * model.y1 + P2 * model.y2)

model.obj = pyo.Objective(rule=obj_function, sense=pyo.maximize)

# STEP 3: Define the constraints

# Budget constraint
def budget_constraint(model):
    return C1 * model.x1 + C2 * model.x2 + P1 * model.y1 + P2 * model.y2 <= B

model.budget_c = pyo.Constraint(rule=budget_constraint)

# Storage capacity constraints
def storage_capacity_constraints(model):
    return [model.x1 <= Q1, model.x2 <= Q2]

model.storage_c1 = pyo.Constraint(rule=storage_capacity_constraints[0])
model.storage_c2 = pyo.Constraint(rule=storage_capacity_constraints[1])

# Active ingredient requirement constraints
def active_ingredient_constraints(model):
    return [E1 * (1 - 0.005) * model.x1 >= R1, E2 * (1 - 0.02) * model.x2 >= R2]

model.active_ingredient_c1 = pyo.Constraint(rule=active_ingredient_constraints[0])
model.active_ingredient_c2 = pyo.Constraint(rule=active_ingredient_constraints[1])

# Manpower and equipment hours constraints
def manpower_equipment_constraints(model):
    return [T1 * model.y1 <= M1, T2 * model.y2 <= M2]

model.manpower_c1 = pyo.Constraint(rule=manpower_equipment_constraints[0])
model.manpower_c2 = pyo.Constraint(rule=manpower_equipment_constraints[1])

# Non-negativity constraints (already defined in the variables)

# Solve the model
solver = pyo.SolverFactory('glpk')
results = solver.solve(model)

# Print the results
print("Optimal profit:", pyo.value(model.obj))
print("Quantity of raw material 1:", pyo.value(model.x1))
print("Quantity of raw material 2:", pyo.value(model.x2))
print("Amount of drug 1 produced:", pyo.value(model.y1))
print("Amount of drug 2 produced:", pyo.value(model.y2))
```
