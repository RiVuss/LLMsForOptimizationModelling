### Code Executability

```python
# Import the necessary libraries
from pyomo.environ import *

# Define the model
model = ConcreteModel()

# Define the parameters
model.C1 = 10000  # Available amount of component C1
model.C2 = 15000  # Available amount of component C2
model.alpha1 = 0.4  # Critical element fraction in component C1
model.alpha2 = 0.2  # Critical element fraction in component C2
model.beta_GCA = 0.3  # Minimum critical element fraction required in GCA
model.beta_GCB = 0.3  # Maximum critical element fraction allowed in GCB
model.R_GCC = 0.3  # Minimum ratio of C1 to C2 in GCC
model.D_GCA = 6000  # Minimum demand for GCA
model.D_GCB = 7000  # Minimum demand for GCB
model.D_GCC = 9000  # Minimum demand for GCC
model.P_GCA = 120  # Profit per liter of GCA
model.P_GCB = 135  # Profit per liter of GCB
model.P_GCC = 155  # Profit per liter of GCC

# Define the variables
model.x1 = Var(within=NonNegativeReals)  # Amount of C1 used to produce GCA
model.x2 = Var(within=NonNegativeReals)  # Amount of C2 used to produce GCA
model.y1 = Var(within=NonNegativeReals)  # Amount of C1 used to produce GCB
model.y2 = Var(within=NonNegativeReals)  # Amount of C2 used to produce GCB
model.z1 = Var(within=NonNegativeReals)  # Amount of C1 used to produce GCC
model.z2 = Var(within=NonNegativeReals)  # Amount of C2 used to produce GCC
model.GCA = Var(within=NonNegativeReals)  # Total amount of GCA produced
model.GCB = Var(within=NonNegativeReals)  # Total amount of GCB produced
model.GCC = Var(within=NonNegativeReals)  # Total amount of GCC produced

# Define the objective function
def obj_rule(model):
    return model.P_GCA * model.GCA + model.P_GCB * model.GCB + model.P_GCC * model.GCC
model.obj = Objective(rule=obj_rule, sense=maximize)

# Define the constraints
def component_availability_C1(model):
    return model.x1 + model.y1 + model.z1 <= model.C1
model.component_availability_C1_constraint = Constraint(rule=component_availability_C1)

def component_availability_C2(model):
    return model.x2 + model.y2 + model.z2 <= model.C2
model.component_availability_C2_constraint = Constraint(rule=component_availability_C2)

def minimum_demand_GCA(model):
    return model.GCA >= model.D_GCA
model.minimum_demand_GCA_constraint = Constraint(rule=minimum_demand_GCA)

def minimum_demand_GCB(model):
    return model.GCB >= model.D_GCB
model.minimum_demand_GCB_constraint = Constraint(rule=minimum_demand_GCB)

def minimum_demand_GCC(model):
    return model.GCC >= model.D_GCC
model.minimum_demand_GCC_constraint = Constraint(rule=minimum_demand_GCC)

def critical_element_fraction_GCA(model):
    return model.alpha1 * model.x1 + model.alpha2 * model.x2 >= model.beta_GCA * model.GCA
model.critical_element_fraction_GCA_constraint = Constraint(rule=critical_element_fraction_GCA)

def critical_element_fraction_GCB(model):
    return model.alpha1 * model.y1 + model.alpha2 * model.y2 <= model.beta_GCB * model.GCB
model.critical_element_fraction_GCB_constraint = Constraint(rule=critical_element_fraction_GCB)

def ratio_C1_C2_GCC(model):
    return model.z1 >= model.R_GCC * model.z2
model.ratio_C1_C2_GCC_constraint = Constraint(rule=ratio_C1_C2_GCC)

def production_quantities_GCA(model):
    return model.GCA == model.x1 + model.x2
model.production_quantities_GCA_constraint = Constraint(rule=production_quantities_GCA)

def production_quantities_GCB(model):
    return model.GCB == model.y1 + model.y2
model.production_quantities_GCB_constraint = Constraint(rule=production_quantities_GCB)

def production_quantities_GCC(model):
    return model.GCC == model.z1 + model.z2
model.production_quantities_GCC_constraint = Constraint(rule=production_quantities_GCC)

# Solve the model
solver = SolverFactory('glpk')
solver.solve(model)

# Print the results
print("Total profit: ", model.obj())
print("Amount of GCA produced: ", model.GCA())
print("Amount of GCB produced: ", model.GCB())
print("Amount of GCC produced: ", model.GCC())
print("Amount of C1 used for GCA: ", model.x1())
print("Amount of C2 used for GCA: ", model.x2())
print("Amount of C1 used for GCB: ", model.y1())
print("Amount of C2 used for GCB: ", model.y2())
print("Amount of C1 used for GCC: ", model.z1())
print("Amount of C2 used for GCC: ", model.z2())
```
