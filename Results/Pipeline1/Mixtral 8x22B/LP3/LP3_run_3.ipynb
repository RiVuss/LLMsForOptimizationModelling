```python
import os
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
from IPython.display import display, Markdown, Latex
from datetime import datetime
```

```python
API_KEY = os.environ['MISTRAL_API_KEY']

MODEL_ID = 'open-mixtral-8x22b'
MODEL_SEED = 3
MODEL_TEMPERATURE = 0.7

file_system_prompt_1 = open("../../system_prompt_1.txt", "r")
file_system_prompt_2 = open("../../system_prompt_2.txt", "r")
SYSTEM_PROMPT_1 = file_system_prompt_1.read()
SYSTEM_PROMPT_2 = file_system_prompt_2.read()
file_system_prompt_1.close()
file_system_prompt_2.close()

FILE_PATH_PROBLEM = '../../../../Datasets/LP_3_Weekly-Production-Metallurgical-Company/'
assert(FILE_PATH_PROBLEM != '../../../../Datasets/')
file_problem_description = open(FILE_PATH_PROBLEM + 'ProblemDescription.txt', 'r')
PROBLEM_DESCRIPTION = file_problem_description.read()
file_problem_description.close()


client = MistralClient(api_key=API_KEY)

print(f'Time of execution: {datetime.now()}')
```

```python
import pyomo.environ as pyo

# Define the model
model = pyo.ConcreteModel()

# Define the parameters
model.c_s1 = 0.30
model.c_ss1 = 0.50
model.c_m1 = 0.20
model.c_a1 = 0.30
model.c_s2 = 0.50
model.c_ss2 = 0.60
model.c_m2 = 0.10
model.c_a2 = 0.20
model.c_s3 = 0.40
model.c_m3 = 0.27
model.c_a3 = 0.20

model.p_1 = 1.50
model.p_2 = 1.80
model.p_3 = 1.97

model.t_s1 = 6
model.t_s2 = 10
model.t_s3 = 8
model.t_m1 = 6
model.t_m2 = 3
model.t_m3 = 8
model.t_a1 = 3
model.t_a2 = 2
model.t_a3 = 2

model.T_s = 8000
model.T_m = 12000
model.T_a = 10000

# Define the decision variables
model.x1 = pyo.Var(domain=pyo.NonNegativeIntegers)
model.y1 = pyo.Var(domain=pyo.NonNegativeIntegers)
model.x2 = pyo.Var(domain=pyo.NonNegativeIntegers)
model.y2 = pyo.Var(domain=pyo.NonNegativeIntegers)
model.x3 = pyo.Var(domain=pyo.NonNegativeIntegers)

# Define the objective function
def obj_rule(model):
    return model.p_1 * (model.x1 + model.y1) - (model.c_s1 * model.x1 + model.c_ss1 * model.y1 + model.c_m1 * model.x1 + model.c_a1 * model.x1) + \
           model.p_2 * (model.x2 + model.y2) - (model.c_s2 * model.x2 + model.c_ss2 * model.y2 + model.c_m2 * model.x2 + model.c_a2 * model.x2) + \
           model.p_3 * model.x3 - (model.c_s3 * model.x3 + model.c_m3 * model.x3 + model.c_a3 * model.x3)

model.obj = pyo.Objective(rule=obj_rule, sense=pyo.maximize)

# Define the constraints
def smelting_time_constraint_p1(model):
    return model.t_s1 * model.x1 <= model.T_s

model.smelting_time_constraint_p1 = pyo.Constraint(rule=smelting_time_constraint_p1)

def smelting_time_constraint_p2(model):
    return model.t_s2 * model.x2 <= model.T_s

model.smelting_time_constraint_p2 = pyo.Constraint(rule=smelting_time_constraint_p2)

def smelting_time_constraint_p3(model):
    return model.t_s3 * model.x3 <= model.T_s

model.smelting_time_constraint_p3 = pyo.Constraint(rule=smelting_time_constraint_p3)

def mechanisation_time_constraint(model):
    return model.t_m1 * model.x1 + model.t_m2 * model.x2 + model.t_m3 * model.x3 <= model.T_m

model.mechanisation_time_constraint = pyo.Constraint(rule=mechanisation_time_constraint)

def assembly_packaging_time_constraint(model):
    return model.t_a1 * model.x1 + model.t_a2 * model.x2 + model.t_a3 * model.x3 <= model.T_a

model.assembly_packaging_time_constraint = pyo.Constraint(rule=assembly_packaging_time_constraint)

# Solve the model
solver = pyo.SolverFactory('glpk')
solver.solve(model)

# Print the results
print("Optimal solution:")
print(f"x1 = {pyo.value(model.x1)}")
print(f"y1 = {pyo.value(model.y1)}")
print(f"x2 = {pyo.value(model.x2)}")
print(f"y2 = {pyo.value(model.y2)}")
print(f"x3 = {pyo.value(model.x3)}")
print(f"Profit = {pyo.value(model.obj)}")
```
