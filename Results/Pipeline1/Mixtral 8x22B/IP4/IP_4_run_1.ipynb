### Code Executability

```python
from pyomo.environ import *

# Parameters (sample data)
rf = 5  # Full-time employee sales rate
rp = 2  # Part-time employee sales rate
St = 5500  # Target sales
Hft = 800  # Target full-time employee hours
Hpt = 320  # Target part-time employee hours
Hfto = 100  # Target full-time overtime hours
Cpt = 3  # Relative cost of part-time overtime hours
Ift = 2  # Relative importance of full-time employee hours
P1, P2, P3, P4 = 1, 1, 1, 1  # Priority factors (sample data)

model = ConcreteModel()

# Decision variables
model.Hf = Var(domain=NonNegativeReals)  # Full-time employee hours
model.Hp = Var(domain=NonNegativeReals)  # Part-time employee hours
model.Hfo = Var(domain=NonNegativeReals)  # Full-time overtime hours
model.Hpo = Var(domain=NonNegativeReals)  # Part-time overtime hours
model.S = Var(domain=NonNegativeReals)  # Sales
model.Ds_minus = Var(domain=NonNegativeReals)  # Negative deviation from sales target
model.Ds_plus = Var(domain=NonNegativeReals)  # Positive deviation from sales target
model.Dft_minus = Var(domain=NonNegativeReals)  # Negative deviation from full-time employee hours target
model.Dft_plus = Var(domain=NonNegativeReals)  # Positive deviation from full-time employee hours target
model.Dfo_minus = Var(domain=NonNegativeReals)  # Negative deviation from full-time overtime hours target
model.Dfo_plus = Var(domain=NonNegativeReals)  # Positive deviation from full-time overtime hours target
model.Dpo_minus = Var(domain=NonNegativeReals)  # Negative deviation from part-time overtime hours target
model.Dpo_plus = Var(domain=NonNegativeReals)  # Positive deviation from part-time overtime hours target

# Objective function
def obj_rule(model):
    return P1 * (model.Ds_minus + model.Ds_plus) + P2 * (model.Dfo_minus + model.Dfo_plus) + P3 * (model.Dft_minus + model.Dft_plus) + Ift * P3 * (model.Dpo_minus + model.Dpo_plus) + Cpt * (model.Dpo_minus + model.Dpo_plus)
model.obj = Objective(rule=obj_rule, sense=minimize)

# Constraints
model.sales_constraint = Constraint(expr=(rf * model.Hf + rp * model.Hp + rf * model.Hfo + rp * model.Hpo - model.Ds_minus + model.Ds_plus == St))
model.full_time_constraint = Constraint(expr=(model.Hf + model.Dft_minus - model.Dft_plus == Hft))
model.full_time_overtime_constraint = Constraint(expr=(model.Hfo + model.Dfo_minus - model.Dfo_plus == Hfto))
model.part_time_constraint = Constraint(expr=(model.Hp + model.Dpo_minus - model.Dpo_plus == Hpt))

# Solve the model
solver = SolverFactory('glpk')
solver.solve(model)

# Print results
print("Full-time employee hours:", model.Hf.value)
print("Part-time employee hours:", model.Hp.value)
print("Full-time overtime hours:", model.Hfo.value)
print("Part-time overtime hours:", model.Hpo.value)
print("Sales:", model.S.value)
print("Negative deviation from sales target:", model.Ds_minus.value)
print("Positive deviation from sales target:", model.Ds_plus.value)
print("Negative deviation from full-time employee hours target:", model.Dft_minus.value)
print("Positive deviation from full-time employee hours target:", model.Dft_plus.value)
print("Negative deviation from full-time overtime hours target:", model.Dfo_minus.value)
print("Positive deviation from full-time overtime hours target:", model.Dfo_plus.value)
print("Negative deviation from part-time overtime hours target:", model.Dpo_minus.value)
print("Positive deviation from part-time overtime hours target:", model.Dpo_plus.value)
```
